```python
import pandas as pd
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
data = pd.read_csv('marks_prediction_data.csv')
```

```python
data.head(1)
```

```python
X = data.drop(['Name', 'Final Exam Score'], axis=1)
Y = data['Final Exam Score']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=1)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
X_test
```

```python
Y_pred = regressor.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred)
```

```python
print ("Mean Squared Error:", mse)
```

Joblib is a set of tools to provide lightweight pipelining in Python. Joblib is optimized to be fast and robust on large data in particular and has specific optimizations for numpy arrays

```python
import joblib
```

In Python, we work with high-level data structures such as lists, tuples, and sets. However, when we want to store these objects in memory, they need to be converted into a sequence of bytes that the computer can understand. This process is called serialization.

The next time we want to access the same data structure, this sequence of bytes must be converted back into the high-level object in a process known as deserialization. Python Pickle library for serialization

Object serialization avoids rerunning  Python code every time you need to access a previously created data frame, variable, or machine learning model


It is the process of storing a data structure in memory so that you can load or transmit it when required without losing its current state.

https://www.datacamp.com/tutorial/pickle-python-tutorial

dump(python object, file) : Persist an arbitrary Python object into one file.
Returns filenames: list of strings     The list of file names in which the data is stored.

```python
#Save the trained model to a file
joblib.dump(regressor, 'my_model.pkl') 
```

```python
#Reconstruct a Python object from a file persisted with joblib.dump
## Load the trained model from a file
regressor1 = joblib.load('my_model.pkl')
```

```python
# Load new test data
new_data = pd.read_csv('new_marks_prediction_data.csv')
new_data
```

```python
#Preprocessing new test data
X_new = new_data.drop(['Final Exam Score'], axis=1)
```

```python
# Make predictions on the new test data using the trained model
y_pred = regressor1.predict(X_new)
```
